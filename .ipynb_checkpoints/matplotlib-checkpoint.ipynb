```python
import pandas as pd
import numpy as numpy
import matplotlib.pyplot as plt
x=['Monday','Tuesday','Wednesday','Thursday','Friday']
y=[23,53,43,43,42]
plt.plot(x,y)
plt.xlabel('Week Days')
plt.ylabel('Sales')
plt.show()
```

```python
w=['Preyas','Maitri','Vedika','Devansh','Shruti','Krushn']
z=[1900,2400,1500,1950,1000,3000]
plt.plot(w,z,color='darkgreen',linestyle='--',linewidth='1',marker='o',label='2025 Sales')
plt.legend()
plt.grid(linestyle=':',color='black')
plt.ylim(0,3500)
plt.show()
```

```python
import pandas as pd
import numpy as numpy
import matplotlib.pyplot as plt
w=['Preyas','Maitri','Vedika','Devansh','Shruti','Krushn']
z=[1900,2400,1500,1950,1000,3000]
plt.bar(w,z)
plt.xlabel('Name')
plt.ylabel('Number')
plt.show()
```

```python
import matplotlib.pyplot as plt
w=['Preyas','Maitri','Vedika','Devansh','Shruti','Krushn']
z=[1900,2400,1500,1950,1000,3000]
plt.pie(z,labels=w,autopct='%1f',colors=['Gold','Silver','Blue','Pink','Green','skyblue'])
plt.show()
```

```python
import matplotlib.pyplot as plt

z = [1900, 2400, 1500, 1950, 1000, 3000]

plt.hist(z, bins=5, edgecolor='black')  # group values into 5 ranges
plt.xlabel("Earnings")
plt.ylabel("Frequency")
plt.title("Distribution of Earnings")
plt.show()
```

```python
import matplotlib.pyplot as plt
w=[1,2,3,4,5,6]
z=[1900,2400,1500,1950,1000,3000]
plt.scatter(w,z)
plt.grid()
plt.show()
```
