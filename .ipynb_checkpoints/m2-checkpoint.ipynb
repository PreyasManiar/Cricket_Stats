```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
score=pd.read_excel("D:\\Desktop\\Python\\Matplotlib\\Cricket_1_2.xlsx")
score.head()
```

```python
w=list(range(0,25,2))
x=score['Overs']
z=score['Runs']
plt.xlabel('Overs')
plt.ylabel('Runs')
plt.grid(True)
plt.yticks(w)
plt.plot(x,z)
plt.show()
```
