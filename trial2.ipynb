```python
#MVP of the match
df.columns = df.columns.str.strip()
df.rename(columns={' ': 'Balls'}, inplace=True)

legal_df = df[df['Outcome'] != 'Wide Ball']

runs = legal_df.groupby('Batsman')['Runs'].sum()
balls = legal_df.groupby('Batsman').size()

strike_rate = (runs / balls * 100).round(2)

mvp_score = (runs + (strike_rate * 0.5)).round(2)

mvp_df = pd.DataFrame({
    'Runs': runs,
    'Balls Faced': balls,
    'Strike Rate': strike_rate,
    'MVP Score': mvp_score
})

mvp_df = mvp_df.sort_values(by='MVP Score', ascending=False)

print("ðŸ† MVP of the Match:")
print(mvp_df.head(1))
```
